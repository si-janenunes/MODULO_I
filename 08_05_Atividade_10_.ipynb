```python
# EXEMPLO SOMA EM SEQUÊNCIA
for i in range (1,11):
  resultado = 3 + i
  print ( "3 +", i , "=", resultado)
```

```python
# EXEMPLO SUBITRAÇÂO EM SEQUÊNCIA

for i in range (1,11):
  resultado = 3 - i
  print ( " 3 -", i , "=", resultado)
```

```python
# EXEMPLO DIVISÂO EM SEQUÊNCIA

for i in range (1,11):
  resultado = 3 / i
  print ( " 3 %" , i , "=", resultado)
```

```python
# EXEMPLO DIVISÂO EM SEQUÊNCIA

for i in range (1,11):
  resultado = 3 * i
  print ( " 3 x", i , "=", resultado)
```
