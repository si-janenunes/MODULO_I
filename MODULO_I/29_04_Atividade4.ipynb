```python
#EXEMPLO 7
# Raízes de Equação quadrática: a.x^2 + b.x + c = 0
# Ex: x^2 - 6x + 10 = 0

a,b,c = 1, -1,5
delta = b**2 - 4*a*c
if delta > 0:
    r1 = -b/2/a + (delta**0.5)/2/a
    r2 = -b/2/a - (delta**0.5)/2/a
    print(" Duas raízes distintas: " , r1, r2)
elif delta == 0:
      r1 = -b/2/a
      r2 = r1
      print("Duas raízes reais iguais: " , r1, r2)
else:
    r1 = -b/2/a + (delta**0.5) /2/a
    r2 = -b/2/a - (delta**0.5) /2/a
    print("Duas raízes complexas conjugadas: " ,r1, r2)
print('Fim')
```
