```python
curso = []
nome = input('Nome do Aluno: ')
while nome != 'fim':
    nota = float(input('Nota do Aluno: '))
    curso.append(nome)
    curso.append(nota)
    nome = input('Nome do Aluno: ')
print(curso)
print ('\b\b')
print ('Total de Alunos= ',len (curso))
```

```python
#Lista de fatiamento em Ordenação

listadesc = list(range(0,20,2))   #Resultado da Função, Valor
print(listadesc)                  #Acesso de todos os Dados
print(listadesc[0:3])                 #Acesso do 1* e ao 3* elemento
print(listadesc[2:-2])                  #Acesso do 3* ao penultimo
```

```python
#Soma de Elementos da lista (Loop)

pesos = [56,73,92,32,45]   #Lista de Valores homogênea = Valores
soma = 0
for peso in pesos:         #Operação Eeuivalente Soma = Soma + Peso
  soma += peso
  print('Soma dos Pesos: ', soma)
# Programa Paytônico
  print('Soma dos pesos: ', sum(pesos))
```

```python
# Somando valores em lista utilizando dados estatísticos

vendas = [100, 50, 80 ,200]          #Lista com valores
#Estatisticas Básicas
media = sum (vendas) / len(vendas)
maior = max (vendas)
menor = min (vendas)

print(f'Média das Vendas: {media:.2f}') #Média das vendas(todos os valores)
print(f'Maiores Vendas: {maior}')       # Maior valor
print(f'Menores vendas: {menor}')       #Menor valor
```
