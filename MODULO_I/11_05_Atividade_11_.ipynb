# //  EXEMPLOS DE LISTAS
Listas são estruturas usadas para guardar vários valores em
uma única variável. Elas são mutáveis, ou seja, podem ser
alteradas após a criação. São muito úteis quando queremos
armazenar dados em ordem, como nomes de alunos, notas
ou produtos.

```python
# //  Exemplo de Lista

alunos = ["Ana", "Bruno", "Carla"]
print (alunos)
```

```python
# Exemplo 1-  Lista Adicionar dado com Append

alunos = ["Ana", "Bruno","Carla"]
print(alunos[0])
alunos.append("Caio")
print(alunos)
```

```python
# Exemplo 2 - Lista Mover ou Alterar dado com POP

alunos = ["Ana", "Bruno","Carla", "Caio"]
print(alunos[3])
item = alunos.pop(0)
alunos.insert (3,item)
print(alunos)
```

```python
#Exemplo 3 -  Lista Deletar dado alterado

alunos = ["Bruno","Carla", "Caio", "Ana"]
print(alunos[0])
del (alunos[1])
print(alunos)
```

```python
#Exemplo 4 -  Lista Remover dado alterado

alunos = ["Bruno","Carla", "Caio", "Ana"]
print(alunos[0])
alunos.remove ("Carla")
print(alunos)
```

```python
# Exemplo 5 - Tamanho da Lista com len

alunos = ["Ana", "Bruno","Caio"]
print(len(alunos))
```

```python
# Exemplo Tamanho da Lista com len

for aluno in alunos :
  print(aluno)
```
