```python
# Exemplo simples de while com salário

# Salário inicial
salario = 1500

# Meta que queremos alcançar
meta = 2000

# Enquanto o salário for menor que a meta
while salario < meta:
    # Mostra o salário antes do aumento
    print(f"Salário antes do aumento: R$ {salario}")

    # Soma 100 ao salário
    salario = salario + 100

    # Mostra o salário depois do aumento
    print(f"Salário depois do aumento: R$ {salario}")
    print("-" * 30)

# Quando sair do while, significa que a meta foi atingida
print("Meta atingida!")
print(f"Salário final: R$ {salario}")
```

```python
# Pede para a pessoa digitar o salário inicial
salario = float(input("Digite o salário inicial: R$ "))

# Define a meta
meta = 2000

# Enquanto o salário for menor que a meta, repete
while salario < meta:
    print(f"Salário antes do aumento: R$ {salario:.2f}")

    # Soma 100 ao salário
    salario = salario + 100

    print(f"Salário depois do aumento: R$ {salario:.2f}")
    print("-" * 30)

# Quando sair do while, a meta foi atingida
print("Meta atingida!")
print(f"Salário final: R$ {salario:.2f}")
```

```python
# Pede ao usuário para digitar o número do cartão
numero = input("Digite o número do cartão de crédito: ")

# Remove espaços e hífens, caso a pessoa digite assim: 1234 5678-9999
numero = numero.replace(" ", "").replace("-", "")

# Variável que vai guardar a soma dos cálculos
soma = 0

# Variável booleana para alternar a regra do algoritmo
alternar = False

# Percorre os números de trás para frente
for digito in numero[::-1]:
    # Converte o caractere digitado em número inteiro
    n = int(digito)

    # Se a flag alternar estiver ativa, dobra o valor
    if alternar:
        n = n * 2

        # Se o resultado passar de 9, subtrai 9
        if n > 9:
            n = n - 9

    # Soma o valor final
    soma = soma + n

    # Troca a flag para o próximo número
    alternar = not alternar

# Verifica se o número é válido
if soma % 10 == 0:
    print("Cartão válido.")
else:
    print("Cartão inválido.")
```

```python
# Saldo inicial do cartão
saldo = 500.00

# Limite total do cartão
limite = 1500.00

# Calcula o saldo disponível somando saldo + limite
saldo_disponivel = saldo + limite

# Mostra os valores na tela
print(f"Saldo em aberto: R$ {saldo:.2f}")
print(f"Limite do cartão: R$ {limite:.2f}")
print(f"Total disponível para uso: R$ {saldo_disponivel:.2f}")

# Pede um valor de compra
compra = float(input("Digite o valor da compra: R$ "))

# Verifica se ainda há limite suficiente
if compra <= saldo_disponivel:
    saldo_disponivel = saldo_disponivel - compra
    print("Compra aprovada!")
    print(f"Novo total disponível: R$ {saldo_disponivel:.2f}")
else:
    print("Compra recusada. Limite insuficiente.")
```
