# O Cenário: Sistema de Triagem de Cargas
Você foi contratado para criar o "Cérebro" de um sistema de triagem para um terminal
logístico em Manaus. O sistema deve processar um lote de 5 mercadorias e decidir,
com base no peso, se elas seguem para a "Rota Norte" ou "Rota Sul".

# Requisitos Obrigatórios:
1. Entrada de Dados: O algoritmo deve solicitar ao usuário o Nome da Carga
(Texto) e o Peso (Real).
2. Processamento (Regra de Decisão): * Se o Peso for maior ou igual a 100kg,
exiba: "Carga Pesada - Rota Norte".
o Se o Peso for menor que 100kg, exiba: "Carga Leve - Rota Sul".
3. Estrutura de Repetição: O sistema deve repetir esse processo exatamente 5
vezes (utilizando PARA ou ENQUANTO).
4. Saída de Dados: Ao final das 5 iterações, o sistema deve exibir a mensagem:
"Lote Processado com Sucesso".

# FLUXOGRAMA

```python
for i in range (5):
  nome= str (input (" Digite o Nome da Carga = "))
  peso= float (input(" Digite o Peso da Carga em kg =  "))
  if peso >=100:
    print(f" {nome} - Carga Pesada - Rota Norte")
  else:
    print(f"{nome} - Carga Leve - Rota Sul")
print("Lote Processado com Sucesso")
```

```python
|   | Cont  | | Nome  | Peso | Processamento| Condição (peso >=) | Saída                    |
|---|-------| ------- | ---- | -------------| -------------------| -------------------------|
| 1 |   0   | Areia   | 100  |  if          |verdadeiro          | Carga Pesada - Rota Norte|
| 2 |   1   | Seixo   | 50   |  else        |falso               | Carga Leve - Rota Sul    |
| 3 |   2   | Ferro   | 900  |  if          |verdadeiro          | Carga Pesada - Rota Norte|
| 4 |   3   | Cimento | 600  |  if          |verdadeiro          | Carga Pesada - Rota Norte|
| 5 |   4   | Cola    | 50   |  else        |falso               | Carga Leve - Rota Sul    |

```
