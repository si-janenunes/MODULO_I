```python
# Interrupção do laço com `break'

print('-- Obs,: digite palavra para encerrar -- ')
while True:
    cor = input('Entre com a palavra:')
    if cor.lower() == ' fechar':
        break
    cor = input('Entre com a cor:')
    if cor.lower()!= 'Branco':
        break
```

```python
# Interrupção do laço com 'break'
print('-- Obs.: digite palavra fim --')
while True:
    cor = input('Entre com a palavra:')
    if cor.lower() == 'fim':
        break
    cor = input('Entre com a palavra:')
    if cor.lower() == 'fim':
        break
```

```python
# while com break - números aleatórios não múltiplos de 3
import random
while True:
    x = random.randint(0,999)
    if(x % 3 == 0):
        break
    print(x, end = ' ')
```

```python
# mostra as entradas do usuário até ele digite 'fim', desconsiderando entradas com # no início
while True:
    entrada = input('entrada> ')
    if entrada[0] == '#':
        continue
    if entrada == '1599':
        break
    print(entrada)
print('Finalização.')
```

```python
# QUESTÃO 1
soma = 0
qtd = 0
while True:
  numero_str = input("Digite números pares ou 'fim'. ").lower()
  if numero_str == 'fim':
      break
  try:
    numero = float (numero_str)

    if numero % 2 == 0:
      soma += numero
      qtd +=1
    else:
      print ("Erro: Digite apenas números pares")
  except ValueError:
      print("Erro: Digite apenas números válidos (ou 'fim'). ")
if qtd >0:
      print(f"Soma: {soma}/ Qtd: {qtd}/ Média: {soma/qtd:.2f}")
```
