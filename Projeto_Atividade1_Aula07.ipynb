# 1. Importando as bibliotecas necessárias

Vamos começar importando as bibliotecas que precisaremos para este exemplo: numpy para operações numéricas, make_classification do sklearn.datasets para gerar dados sintéticos, train_test_split para dividir os dados, LogisticRegression para o modelo e accuracy_score para avaliação.

```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

print("Bibliotecas importadas com sucesso!")
```

# 2. Gerando um conjunto de dados sintético
Para este exemplo, vamos criar um conjunto de dados binário de classificação com 1000 amostras e 20 características, onde 2 delas são informativas.

```python
X, y = make_classification(n_samples=1000, n_features=20, n_informative=2, n_redundant=10, random_state=42)

print(f"Formato das características (X): {X.shape}")
print(f"Formato do target (y): {y.shape}")
print("Primeiras 5 amostras de X:\n", X[:5])
print("Primeiras 5 amostras de y:\n", y[:5])
```

# 3. Dividindo os dados em conjuntos de treinamento e teste
É uma prática padrão dividir os dados em um conjunto de treinamento (para o modelo aprender) e um conjunto de teste (para avaliar o desempenho do modelo em dados não vistos).

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print(f"Formato de X_train: {X_train.shape}")
print(f"Formato de X_test: {X_test.shape}")
print(f"Formato de y_train: {y_train.shape}")
print(f"Formato de y_test: {y_test.shape}")
```

# 4. Treinando um modelo de Regressão Logística
Vamos usar um modelo de Regressão Logística, que é um classificador linear simples e eficaz, para treinar nossos dados.

```python
model = LogisticRegression(random_state=42)
model.fit(X_train, y_train)

print("Modelo de Regressão Logística treinado com sucesso!")
```

# 5. Fazendo previsões no conjunto de teste
Agora que o modelo foi treinado, podemos usá-lo para fazer previsões nos dados que ele nunca viu antes (o conjunto de teste).

```python
y_pred = model.predict(X_test)

print("Primeiras 10 previsões:\n", y_pred[:10])
print("Primeiros 10 valores reais:\n", y_test[:10])
```

# 6. Avaliando o desempenho do modelo
Finalmente, avaliaremos o desempenho do nosso modelo usando a métrica de acurácia, que mede a proporção de previsões corretas.

```python
accuracy = accuracy_score(y_test, y_pred)

print(f"Acurácia do modelo: {accuracy:.4f}")
```
